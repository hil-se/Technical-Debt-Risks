</h1> Pre- processing for embeddings </h1>

```python
comments=data[['Abstract']]
```

```python
labels=data[['label']]
```

```python
labels_list=[]
for i, row in labels.iterrows():
    labels_list.append(row['label'])
    
comments_list=[]
for i, row in comments.iterrows():
    comments_list.append(row['Abstract'])
```

```python
comments_list
```

```python
labels_list
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
NUM_WORDS=20000
tokenizer = Tokenizer(num_words=NUM_WORDS,filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n\'',
                      lower=True)
tokenizer.fit_on_texts(comments_list)
X = tokenizer.texts_to_sequences(comments_list)
X,l= np.array(X), np.array(labels)
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

```python
X
```

```python
l
```

```python
label_encoder = LabelEncoder()
vec = label_encoder.fit_transform(l)
y=to_categorical(vec)
```

```python
y
```

```python
X = pad_sequences(X)
```

```python
X
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=1)
data = {}
data["X_train"] = X_train
data["X_val"]= X_val
data["y_train"] = y_train
data["y_val"] = y_val
data["tokenizer"] = tokenizer
data["int2label"] =  {0: "no", 1: "yes"}
data["label2int"] = {"no": 0, "yes": 1}
    
```

```python
print('Shape of X train and X validation tensor:', X_train.shape,X_val.shape)
```

```python
print('Shape of label train and validation tensor:', y_train.shape,y_val.shape)
```

```python
import gensim
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess

from gensim.models.keyedvectors import KeyedVectors
```

```python
word_vectors = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)
```

```python
word_vectors
```

```python
EMBEDDING_DIM=300
vocabulary_size=min(len(word_index)+1,NUM_WORDS)
embedding_matrix = np.zeros((vocabulary_size, EMBEDDING_DIM))
for word, i in word_index.items():
    if i>=NUM_WORDS:
        continue
    try:
        embedding_vector = word_vectors[word]
        embedding_matrix[i] = embedding_vector
    except KeyError:
        embedding_matrix[i]=np.random.normal(0,np.sqrt(0.25),EMBEDDING_DIM)
```

```python
embedding_matrix
```

```python
from tensorflow.keras.layers import Embedding
embedding_layer = Embedding(vocabulary_size,
                            EMBEDDING_DIM,
                            weights=[embedding_matrix],
                            trainable=False)
```

```python
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Embedding
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Dense, Embedding, Conv2D, MaxPooling2D, Dropout,concatenate,GlobalMaxPooling1D
from tensorflow.keras.layers import Reshape, Flatten
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Model
from tensorflow.keras import regularizers
sequence_length = X_train.shape[1]
filter_sizes = [3,4,5]
num_filters = 100
drop = 0.5
```

```python
sequence_length
```

<h2> Define CNN with simple one convolutional layer and one pooling layer </h2>

```python
model = Sequential()

embedding_layer = Embedding(vocabulary_size,
                            EMBEDDING_DIM,
                            weights=[embedding_matrix],
                            trainable=False)
model.add(embedding_layer)

model.add(Conv1D(128, 5, activation='sigmoid'))
model.add(GlobalMaxPooling1D())
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
```

```python
print(model.summary())
```

```python
history = model.fit(X_train, y_train, batch_size=128, epochs=6, verbose=1, validation_split=0.2)
score = model.evaluate(X_val, y_val, verbose=1)
```

```python
print("Test Score:", score[0])
print("Test Accuracy:", score[1])
```
